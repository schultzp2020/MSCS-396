```python
import numpy as np
```

```python
def RC4_K(key, n=256):
  S = np.array(range(n), dtype=int)
  j = 0
  for i in range(n):
    j = (j + S[i] + key[i % len(key)]) % n
    S[i], S[j] = S[j], S[i]
  return S
```

```python
def RC4_PRGA(length, S, n=256):
  key_stream = np.zeros(length, dtype=int)
  i = 0
  j = 0
  for k in range(length):
    i = (i + 1) % n
    j = (j + S[i]) % n
    S[i], S[j] = S[j], S[i]
    key_stream[k] = S[(S[i] + S[j]) % n]
  return key_stream
```

```python
RC4_K([1, 1, 3])
```

```python
RC4_PRGA(1000, RC4_K([1, 1, 3]))
```

```python
RC4_K([1, 2, 1], 4)
```
