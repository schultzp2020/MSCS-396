```python
def extended_euclidean(m, n):
  if n == 0:
    return 1, 0, m
  x, y, g = extended_euclidean(n, m % n)
  return y, x - (m // n)*y, g
```

```python
def pollard_rho(n):
  f = lambda x_: x_**2 + 1
  x = 2
  y = 2
  g = 1
  while g == 1:
    x = f(x) % n
    y = f(f(y)) % n
    _, _, g = extended_euclidean(abs(x-y), n)
  return g
```
